```python
import yt
yt.enable_plugins()
ds = yt.load("galaxy_properties/DD0125/output_0125")
```

```python
center1 = ds.arr([0.32250, 0.34228, 0.34496], 'Mpccm/h')
center2 = ds.arr([0.33480, 0.34586, 0.33340], 'Mpccm/h')
center3 = ds.arr([0.34894, 0.34557, 0.34687], 'Mpccm/h')
radius1 = ds.quan(7.87, 'kpccm/h')
radius2 = ds.quan(9.23, 'kpccm/h')
radius3 = ds.quan(6.89, 'kpccm/h')

x_density = yt.ProjectionPlot(ds,"x",["density","temperature"],width=(20,"kpc"),weight_field="density")
x_density["density"]
```

```python
sph1 = ds.sphere(center1, radius1)
sph2 = ds.sphere(center2, radius2)
sph3 = ds.sphere(center3, radius3)
```

```python
#ad = ds.all_data()
hot1 = sph1["temperature"].in_units('K') > 1e6
warm1 = (sph1["temperature"].in_units('K') > 1e4) & (sph1["temperature"].in_units('K') < 1e6)
cool1 = (sph1["temperature"].in_units('K') < 1e4) & (sph1["temperature"].in_units('K') > 1e2)
cold1 = sph1["temperature"].in_units('K') < 1e2

hot2 = sph2["temperature"].in_units('K') > 1e6
warm2 = (sph2["temperature"].in_units('K') > 1e4) & (sph2["temperature"].in_units('K') < 1e6)
cool2 = (sph2["temperature"].in_units('K') < 1e4) & (sph2["temperature"].in_units('K') > 1e2)
cold2 = sph2["temperature"].in_units('K') < 1e2

hot3 = sph3["temperature"].in_units('K') > 1e6
warm3 = (sph3["temperature"].in_units('K') > 1e4) & (sph3["temperature"].in_units('K') < 1e6)
cool3 = (sph3["temperature"].in_units('K') < 1e4) & (sph3["temperature"].in_units('K') > 1e2)
cold3 = sph3["temperature"].in_units('K') < 1e2

#cutregions
hot_1 = sph1.cut_region(['obj["temperature"] > 1e6'])
warm_1 = sph1.cut_region(['(obj["temperature"] < 1e6) & (obj["temperature"] > 1e4)'])
cool_1 = sph1.cut_region(['(obj["temperature"] < 1e4) & (obj["temperature"] > 1e2)'])
cold_1 = sph1.cut_region(['obj["temperature"] < 1e2'])

hot_2 = sph2.cut_region(['obj["temperature"] > 1e6'])
warm_2 = sph2.cut_region(['(obj["temperature"] < 1e6) & (obj["temperature"] > 1e4)'])
cool_2 = sph2.cut_region(['(obj["temperature"] < 1e4) & (obj["temperature"] > 1e2)'])
cold_2 = sph2.cut_region(['obj["temperature"] < 1e2'])

hot_3 = sph3.cut_region(['obj["temperature"] > 1e6'])
warm_3 = sph3.cut_region(['(obj["temperature"] < 1e6) & (obj["temperature"] > 1e4)'])
cool_3 = sph3.cut_region(['(obj["temperature"] < 1e4) & (obj["temperature"] > 1e2)'])
cold_3 = sph3.cut_region(['obj["temperature"] < 1e2'])
```

```python
print("Sphere 1 cool")
print(cool_1.quantities.total_mass()[0] / sph1.quantities.total_mass()[0])
print("Sphere 2 cool")
print(cool_2.quantities.total_mass()[0] / sph2.quantities.total_mass()[0])
print("Sphere 2 warm")
print(warm_2.quantities.total_mass()[0] / sph2.quantities.total_mass()[0])
print("Sphere 3 cool")
print(cool_3.quantities.total_mass()[0] / sph3.quantities.total_mass()[0])
print("Sphere 3 warm")
print(warm_3.quantities.total_mass()[0] / sph3.quantities.total_mass()[0])
```
